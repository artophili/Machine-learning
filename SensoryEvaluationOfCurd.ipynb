```python
import pandas as pd
import numpy as np
from scipy.stats import f_oneway
```

```python
#Cashew curd
mouth_feel = [3,4,2,2,2,3,3,2,2,3]
texture = [2,3,2,2,2,2,3,2,3,3]
taste = [3,3,2,2,3,4,3,3,3,3]
smell = [4,3,2,1,2,4,3,3,2,3]
color = [2,3,2,2,2,2,3,3,2,3]
acceptability = [3,3,2,2,2,3,3,3,2,3]
```

```python
f_stats, p_value = f_oneway(mouth_feel,texture,taste,smell,color,acceptability)
```

```python
alpha = 0.05

if p_value < 0.05:
  print('reject null and accept alternate')
else:
  print('accept null there is no differnce')
```

```python
def sme(l1):
  sd = tstd(l1)
  s = sd/3.166
  mean = np.mean(l1)
  return mean, s
```

```python
a = sme(texture)
print(a)
```
