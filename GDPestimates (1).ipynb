```python
from statsmodels.tsa.stattools import adfuller

def adfuller_test(WPI):
  result = adfuller(WPI)
  lables = ['ADF Test stats', 'p-value', 'lags used', 'no of obs used']
  for value, label in zip(result, lables):
    print(label + ':' + str(value))
  if result[1] <= 0.05:
    print('Data is stationary')
  else:
    print('data is not stationary')
```

```python
adfuller_test(df['1. Agriculture, forestry & fishing'])
```

```python
selected_col = df['1. Agriculture, forestry & fishing']

agr_data = pd.DataFrame(selected_col)
```

```python
agr_data.head()
```

```python
agr_data['seasonal_first_diff'] = agr_data['1. Agriculture, forestry & fishing'] - agr_data['1. Agriculture, forestry & fishing'].shift(4)
```

```python
agr_data.head(10)
```

```python
adfuller_test(agr_data['seasonal_first_diff'].dropna())
```

```python
#d = 1
plt.figure(figsize = (10,6))
plt.plot(agr_data.index, agr_data['seasonal_first_diff'], marker='o',linestyle='-')
plt.title('Stationary GDP data of Agriculture, forestry & fishing')
plt.xlabel('Time')
plt.ylabel('GDP quarter wise data of Agriculture, forestry & fishing')
plt.grid(True)
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
fig = plt.figure(figsize = (12,8))
ax1 = fig.add_subplot(211)
fig = plot_acf(agr_data['seasonal_first_diff'].dropna(),lags=30,ax = ax1)
ax2 = fig.add_subplot(212)
fig = plot_pacf(agr_data['seasonal_first_diff'].dropna(),lags=15,ax = ax2)
```

```python
'''p = 2
q = 3
d = 1'''

from statsmodels.tsa.statespace.sarimax import SARIMAX

model = SARIMAX(agr_data['1. Agriculture, forestry & fishing'], order = (2, 1, 3), seasonal_order=(2,1,3,4))
results = model.fit()
```

```python
num_periods = 12
agr_data['forecast'] = results.predict(start = len(agr_data), end = len(agr_data) + num_periods-1, dynamic = True)
agr_data[['1. Agriculture, forestry & fishing', 'forecast']].plot(figsize=(12,8))
```

```python
agr_data['forecast'] = results.predict(start = 80, end = 100, dynamic = True)
agr_data[['1. Agriculture, forestry & fishing', 'forecast']].plot(figsize=(12,8))
```
