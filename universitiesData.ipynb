```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_2023 = pd.read_excel('/content/2023.xlsx')
df_2023.head()
```

```python
# @title Top 5 Institutes by Score

df_2023.sort_values(by='Rank', ascending=True).head(5).plot.bar(x='Institute Name', y='Score')
```

```python
# @title Score by State

import matplotlib.pyplot as plt
df_2023.groupby('State')['Score'].mean().plot.bar()
plt.xlabel('State')
_ = plt.ylabel('Average Score')
```
