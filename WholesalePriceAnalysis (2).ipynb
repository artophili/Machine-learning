```python
import matplotlib.pyplot as plt

year_data = df['Year'].tolist()
cashew_nut_data = df['Cashew nut'].tolist()
#Time series of WPI for Cashew nut
plt.figure(figsize=(10, 6))
plt.plot(year_data, cashew_nut_data, marker='o', linestyle='-')
plt.title('Wholesale Price Index (WPI) for Cashew nut Over Time')
plt.xlabel('Year')
plt.ylabel('WPI For Cashew Nut')
plt.grid(True)
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
df.fillna(0, inplace=True)


# Perform decomposition analysis on the WPI data
decomposition = seasonal_decompose(df['Cashew nut'], model='additive', period=6)

# Plot the decomposed components
plt.figure(figsize=(10, 8))
plt.subplot(4, 1, 1)
plt.plot(cashew_nut_data, decomposition.observed, label='Observed')
plt.legend()

plt.subplot(4, 1, 2)
plt.plot(cashew_nut_data, decomposition.trend, label='Trend')
plt.legend()

plt.subplot(4, 1, 3)
plt.plot(cashew_nut_data, decomposition.seasonal, label='Seasonal')
plt.legend()

plt.subplot(4, 1, 4)
plt.plot(cashew_nut_data, decomposition.resid, label='Residual')
plt.legend()

plt.suptitle('Decomposition Analysis of WPI Data')
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller

result = adfuller(df['Cashew nut'], autolag='AIC')
print('ADF Statistic:', result[0])
print('p-value:', result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t{}: {}'.format(key, value))


df['Cashew nut'].plot()
'''#Time series is non stationary because p-value is more than 0.05

df['Differenced_Data'] = df['Cashew nut'].diff().dropna()
df['Differenced_Data'] = df['Cashew nut'].diff().diff().dropna()'''
```

```python
from statsmodels.tsa.stattools import adfuller

test_result = adfuller(df['Apple'])
```

```python
def adfuller_test(WPI):
  result = adfuller(WPI)
  lables = ['ADF Test stats', 'p-value', 'lags used', 'no of obs used']
  for value, label in zip(result, lables):
    print(label + ':' + str(value))
  if result[1] <= 0.05:
    print('Data is stationary')
  else:
    print('data is not stationary')
```

```python
adfuller_test(df['Apple'])
```

```python
selected_col = df['Apple']
apple_data = pd.DataFrame(selected_col)
```

```python
apple_data.head()
```

```python
apple_data['seasonal_first_diff'] = apple_data['Apple'] - apple_data['Apple'].shift(12)
```

```python
apple_data.head(14)
```

```python
adfuller_test(apple_data['seasonal_first_diff'].dropna())
```

```python
apple_data['seasonal_sec_diff'] = apple_data['seasonal_first_diff'] - apple_data['seasonal_first_diff'].shift(12)
```

```python
apple_data.head(30)
```

```python
adfuller_test(apple_data['seasonal_sec_diff'].dropna())
```

```python
plt.figure(figsize = (10,6))
plt.plot(apple_data.index, apple_data['seasonal_sec_diff'], marker='o',linestyle='-')
plt.title('WPI of Apple data')
plt.xlabel('Time')
plt.ylabel('WPI of Apple')
plt.grid(True)
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
fig = plt.figure(figsize = (12,8))
ax1 = fig.add_subplot(211)
fig = plot_acf(apple_data['seasonal_sec_diff'].dropna(),lags=40,ax = ax1)
ax2 = fig.add_subplot(212)
fig = plot_pacf(apple_data['seasonal_sec_diff'].dropna(),lags=40,ax = ax2)
```

```python
#Selection of p, q ,d
#p is AR model lags
#d differencing
#q MA lags
'''p = 2
q = 6
d = 2'''

from statsmodels.tsa.statespace.sarimax import SARIMAX

model = SARIMAX(apple_data['Apple'], order = (2,2,6), seasonal_order=(2,2,6,12))
results = model.fit()
```

```python
num_periods = 12
apple_data['forecast'] = results.predict(start = len(apple_data), end = len(apple_data) + num_periods-1, dynamic = True)
apple_data[['Apple', 'forecast']].plot(figsize=(12,8))
```
